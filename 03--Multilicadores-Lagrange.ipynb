```python
from sympy import *
from sympy.abc import x, y
from sympy.plotting import plot3d
```

```python
lam = symbols('lambda', real = True)
```

```python
f = 9*pi*x + 3*pi*y
f
```

```python
g = x + root(y**2 + 9, 2) - 21
g
```

```python
L = f - lam * g
```

```python
# Gradientes
gradL = [diff(L,c) for c in [x,y]] 
KKT_eqs = gradL + [g]
KKT_eqs
```

```python
stationary_points = solve(KKT_eqs, [x, y, lam], dict=True)
stationary_points 
```

```python
diff(g, y)
```

```python
solve([9*pi - 2*lam, 3*pi - (lam*y)/root(y**2 + 9, 2)], (lam, y) , dict=True)
```

```python
solve(2*x + sqrt((6*sqrt(5)/5)**2 + 9) - 21, x)
```

```python
[f.subs(p) for p in stationary_points]
```
