```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Graficar espacios vectoriales
<img src="./img/a_vector_field.png" alt="a a_vector_field" />

```python
x, y = np.meshgrid(np.linspace(-5, 5, 10), np.linspace(-5, 5, 10))
a1 = 2*y
a2 = x

plt.quiver(x, y, a1, a2)
plt.show()
```

<img src="./img/b_vector_field.png" alt="a a_vector_field" />

```python
a1 = x/(np.sqrt(x**2 + y**2))
a2 = y/(np.sqrt(x**2 + y**2))

plt.quiver(x, y, a1, a2)
plt.show()
```
